## Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
```

```python
sns.boxplot(x=df['price'], y=df['floors'].sort_values(ascending=True).astype('str'), showfliers=False)
plt.title('Влияние этажности на цену')
plt.xticks(rotation=50);
```

Этажность оказывает небольшое влияние на цену. Наибольшая цена у объектов недвижимости находящимся на 2.5 этаже.

```python
sns.jointplot(x=df['price'], y=df['grade'],  kind='reg', color='green')
plt.title('Влияние состояния конструкций на цену')
```

Состояние конструкций оказывает большое влияние на цену.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'],  kind='reg', color='orange')
plt.title('Влияние жилой площади на цену')
```

Жилая площадь оказывает большое влияние на цену

```python
sns.boxplot(x=df['price'], y=df['view'].sort_values(ascending=True).astype('str'), showfliers=False)
plt.title('Влияние вида из окна на цену')
plt.xticks(rotation=50);
```

Вид из окна (окружающая территория) оказывает значительное влияние на цену
